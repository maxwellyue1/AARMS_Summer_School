```python
re_space = np.linspace(0, 1, 11)
rf_space = np.linspace(0, 0.1, 11)
```

```python
solution = solve_seirx_model(re = 1, rf = 0.01)
t = np.linspace(0, 1000, 1001)
plt.figure(1)
plt.plot(t, solution[:, 0], label = 'S')
plt.plot(t, solution[:, 1], label = 'E')
plt.plot(t, solution[:, 2], label = 'I')
plt.plot(t, solution[:, 3], label = 'R')
#plt.plot(t, solve_seirx_model()[:, 4], label = 'X')
plt.plot(t, 0.5 * np.tanh(0.1 *solution[:, 4]) + 0.5, label = 'transformed x')
plt.legend()
```

```python
for re in re_space: 
    solution = solve_seirx_model(re = re, rf = 0.01)
    plt.plot(t, solution[:, 0], label = 'S')
    plt.plot(t, solution[:, 1], label = 'E')
    plt.plot(t, solution[:, 2], label = 'I')
    plt.plot(t, solution[:, 3], label = 'R')
    #plt.plot(t, solve_seirx_model()[:, 4], label = 'X')
    plt.plot(t, 0.5 * np.tanh(re * solution[:, 4]) + 0.5, label = 'transformed x')
    plt.legend()
    plt.title('re = ' + str(re))
    plt.show()
```
